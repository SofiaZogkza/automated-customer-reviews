```python
import pandas as pd

df_clustering_file = pd.read_csv("../outputs/sofia_product_clusters.csv")
# print(df_clustering_file.head())
print(df_clustering_file.shape)
```

```python
df_classification_file = pd.read_csv("../outputs/data_with_sentiment_clean.csv")
# print(df_clustering_file.head())
print(df_classification_file.shape)
```

```python
import pandas as pd


# Merge the two DataFrames
df_merged = pd.merge(
    df_clustering_file,
    df_classification_file,
    left_on="ProductID",
    right_on="id",
    how="inner"  # you could use "outer" if some IDs are missing
)

# Check the shape of the new DataFrame
print(df_merged.shape)

# Preview the first few rows
print(df_merged.head())
```
